```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = []
model_list = ['open_llama-2-3b', 'llama-2-7b-chat/', 'llama-2-13b-chat']
# model_list = ['open_llama-2-3b']
df = pd.DataFrame()
for model in model_list:
    file_path = '/Users/zyxu/Documents/py/NLP/rethink_icl/out/{}/mean.txt'.format(model)
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split(',')
            parts.append(model)
            data.append(parts)
```

```python
len(data)
```

```python
df = pd.DataFrame(data, columns=['Dataset', 'few_shot', 'Value', 'Model'])
df['Value'] = pd.to_numeric(df['Value'], errors='coerce')

df = df.loc[df['few_shot'] == '16']
df
```

```python
# Split the Dataset column
df['flipped_labels'] = df['Dataset'].str.extract('(\d+)', expand=False)  # Extract numbers
# Map the label info to desired values
label_mapping = {
    'glue-rte_gold_limit1000': '100',
    'glue-rte_permutated_train_labels': '0',
    'glue-sst2_gold_limit1000': '100',
    'glue-sst2_permutated_train_labels': '0',
    'glue-qqp_gold_limit1000': '100',
    'glue-qqp_permutated_train_labels': '0',
    'glue-wnli_gold_limit1000': '100',
    'glue-wnli_permutated_train_labels': '0',
    'subj_gold_limit1000': '100',
    'subj_permutated_train_labels': '0',
}

# If flipped_labels is NaN, then map using the mapping
df['flipped_labels'] = df.apply(lambda row: label_mapping.get(row['Dataset'], row['flipped_labels']), axis=1)
df['Dataset'] = df['Dataset'].str.split('_').str[0]  # Keep only the prefix

df
```

```python
df['flipped_labels'] = df['flipped_labels'].astype(int)
```
